```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
## regression problem statement bcz we are here predicting the salary which is continuous
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
##label encoder
label_encode=LabelEncoder()
df['Gender']=label_encode.fit_transform(df['Gender'])
```

```python
##one hot encode
one_hot=OneHotEncoder(handle_unknown='ignore')
geo_encode=one_hot.fit_transform(df[['Geography']]).toarray()
geo_encode_df=pd.DataFrame(geo_encode,columns=one_hot.get_feature_names_out(['Geography']))
```

```python
geo_encode_df
```

```python
df.head()
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_encode_df],axis=1)
```

```python
df.head()
```

```python
## split our data
y=df['EstimatedSalary']
x=df.drop(['EstimatedSalary'],axis=1)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
## save the encoder and label for later use

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

with open('label_encoder.pkl','wb') as file:
  pickle.dump(label_encode,file)
with open('one_hott.pkl','wb') as file:
  pickle.dump(one_hot,file)
with open('scalerr.pkl','wb') as file:
  pickle.dump(scaler,file)
```

### ANN with regression problem statement

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),## 1st hidden layer with 64 neurons
    Dense(32,activation='relu',),#2nd hidden layer with 32 neurons
    Dense(1)# output layerfor regression


])

## compile the model
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
import datetime
log_dir='regressionlogs/fit/' +datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## set up early stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
## train the model
history=model.fit(
    x_train,y_train,
    validation_data=(x_test,y_test),
    epochs=100,
    callbacks=[early_stopping_callback,tensorflow_callback]
)
```

```python
## evaluate model on the test data
test_loss,test_mae=model.evaluate(x_test,y_test)
print(f'Test MAE :{test_mae}')
```
