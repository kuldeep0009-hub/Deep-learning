```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1) #axis 1 means columns wise
```

```python
df.head()
```

```python
## encode categortical variable
label_encode=LabelEncoder()
df['Gender']=label_encode.fit_transform(df['Gender'])
```

```python
## one hot encode
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder()
geo_encoder=one_hot.fit_transform(df[['Geography']])
geo_encoder
```

```python
df.head()
```

```python
geo_encoder.toarray()
```

```python
one_hot.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=one_hot.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
## combine one hot columns to data
df=pd.concat([df.drop('Geography',axis=1),geo_encoder_df],axis=1)
```

```python
df.head()
```

```python
## save the encoders and scaler
import pickle
with open('label_encode.pkl','wb') as file:
  pickle.dump(label_encode,file)
with open('one_hot.pkl','wb') as file:
  pickle.dump(one_hot,file)
```

```python
## divide the datset into indepemdent and dependent
x=df.drop('Exited',axis=1)
```
